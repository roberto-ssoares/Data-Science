# 📘 1. Business Understanding

## 1.1 Contexto do Negócio

A gestão orientada a dados tem se tornado um diferencial competitivo em organizações que lidam com grandes volumes
de informações sobre pessoas, sejam elas colaboradores ou clientes.

Neste projeto, a empresa RetailX enfrenta dois desafios centrais:

- A alta rotatividade de funcionários, que impacta custos operacionais, prazos e qualidade dos serviços.

- A necessidade de compreender melhor o comportamento dos clientes para otimizar estratégias de CRM.

A adoção de People Analytics permite transformar dados comportamentais e demográficos em insights acionáveis,
apoiando decisões estratégicas baseadas em evidências.

---

## 1.2 Problema de Negócio — RH

Texto sugerido:

- A rotatividade de funcionários representa um problema crítico para a RetailX.
Além dos custos diretos com recrutamento e treinamento, a saída frequente de colaboradores
compromete a continuidade operacional e a retenção de conhecimento.

- O desafio consiste em compreender quais fatores influenciam a decisão de um funcionário deixar a empresa
e estimar, de forma antecipada, o risco de desligamento.

---

## 1.3 Problema de Negócio — CRM

Texto sugerido:

- No contexto de CRM, a RetailX dispõe de dados relacionados ao comportamento de compra,
campanhas de marketing e perfil socioeconômico dos clientes.

- O desafio é identificar padrões de comportamento que permitam a criação de personas,
possibilitando ações mais direcionadas, eficientes e alinhadas às necessidades dos clientes.

---

## 1.4 Objetivos Analíticos

### RH

- Identificar variáveis associadas à alta rotatividade.

- Estimar a probabilidade de saída de funcionários.

- Apoiar estratégias de retenção baseadas em dados.

### CRM

- Identificar segmentos (personas) de clientes.

- Analisar características e comportamentos de cada persona.

- Apoiar decisões de marketing e relacionamento.

---

## 1.5 Perguntas-Chave do Projeto

- Quais fatores estão mais associados à saída de funcionários?

- Por que turnover é um problema financeiro?

- É possível antecipar o risco de desligamento?

- Quanto custa perder um funcionário?

- Quais perfis de clientes apresentam comportamentos semelhantes?

- Como esses insights podem apoiar decisões práticas de RH e CRM?

- Como dados podem reduzir esse risco?

---

## 1.6 Critérios de Sucesso

### Critérios Técnicos

- Modelos interpretáveis e consistentes

- Métricas adequadas ao problema (ex.: AUC, acurácia, silhouette score)

- Contexto do negócio

- Hipóteses iniciais

- KPIs relevantes (turnover rate, churn, LTV)


### Critérios de Negócio

- Clareza dos insights gerados

- Possibilidade de aplicação prática

- Apoio efetivo à tomada de decisão

---

- Ao final da etapa de Business Understanding, o problema está claramente definido, 
os objetivos analíticos estão alinhados às necessidades do negócio e 
os próximos passos do projeto podem ser conduzidos de forma estruturada.
