<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>‚ùó Observe:</strong></font>

<font size="2" color='#66666'>

O trecho de c√≥digo Python **DATA_PATH = next((p for p in candidatos if p.exists()), None)** procura na lista ou iter√°vel **"candidato" so primeiro objeto de caminho que realmente existe** e o atribui √† vari√°vel **DATA_PATH**. 
- Segue abaixo uma descri√ß√£o detalhada de como funciona:
    1. **"candidatos"**
       - Presume-se que seja um iter√°vel (como uma lista) contendo objetos de caminho (provavelmente **pathlib.Path** objetos ou strings representando caminhos de arquivos).
    2. **(p for p in candidatos if p.exists())**
       - Esta √© uma express√£o geradora.
       - Ele itera sobre cada item **(p)** no **"candidatos"** iter√°vel.
       - A **if p.exists()** condi√ß√£o filtra os itens, retornando apenas os caminhos para os quais o **.exists()** m√©todo retorna um valor **True** (ou seja, o arquivo ou diret√≥rio existe no sistema de arquivos).
       - Fundamentalmente, por ser um gerador, ele n√£o cria uma lista completa de caminhos existentes na mem√≥ria; ele os produz um de cada vez, de forma pregui√ßosa.
    3. **next(..., None)**
       - A fun√ß√£o integrada **next()** √© usada para recuperar o primeiro item do gerador.
       - Se o gerador produzir pelo menos um caminho existente, **next()** ele retorna esse primeiro caminho existente e o processo termina.
       - Se o gerador estiver esgotado (ou seja, nenhum dos caminhos em **"candidatos"** existir), a **next()** fun√ß√£o retorna o valor padr√£o fornecido, que √© **None**. 
- Em resumo, a linha de c√≥digo encontra de forma eficiente o primeiro caminho de dados v√°lido/existente em uma lista de candidatos potenciais ou o define **DATA_PATH** como falso **None** caso nenhum caminho v√°lido seja encontrado.

<font size="3" color='#66666'></font></div>

````python
# Converte 'mes' para datetime.
# No notebook de refer√™ncia: format='%Y-%m'
df["mes"] = pd.to_datetime(df["mes"], format="%Y-%m", errors="coerce")

# Valida datas inv√°lidas
n_invalid = df["mes"].isna().sum()
if n_invalid > 0:
    raise ValueError(f"Existem {n_invalid} valores inv√°lidos em 'mes'. Verifique o formato do arquivo.")

# Converte a m√©trica para num√©rico (por seguran√ßa)
df["usuarios_ativos"] = pd.to_numeric(df["usuarios_ativos"], errors="coerce")

# Ordena e seta √≠ndice
df_ts = df.sort_values("mes").set_index("mes")

# Remove duplicatas no √≠ndice (se existirem), agregando por soma (ajuste se fizer mais sentido m√©dia)
if df_ts.index.duplicated().any():
    df_ts = df_ts.groupby(level=0)["usuarios_ativos"].sum().to_frame()

print("""
Convers√£o de datas e defini√ß√£o do √≠ndice temporal:
    - Converte 'mes' para datetime.
    - Valida datas inv√°lidas
    - Converte a m√©trica para num√©rico (por seguran√ßa)
    - Ordena e seta √≠ndice
    - Remove duplicatas no √≠ndice (se existirem), agregando por soma (ajuste se fizer mais sentido m√©dia)
""")
print("Head (10)")
print(df_ts.head(10))
````

<font size="3" color='#66666'></font></div>

````python
# Inferir frequ√™ncia (mensal, semanal, di√°ria etc.)
freq_inferida = pd.infer_freq(df_ts.index)
print("Frequ√™ncia inferida:", freq_inferida)

# Se n√£o conseguir inferir, assumimos mensal (MS) por padr√£o do projeto
freq_alvo = freq_inferida if freq_inferida is not None else "MS"

# Regulariza a s√©rie para frequ√™ncia alvo
serie = df_ts["usuarios_ativos"].asfreq(freq_alvo)

print("Total de pontos:", len(serie))
print("Missing ap√≥s asfreq:", int(serie.isna().sum()))

# Estrat√©gia de preenchimento (simples e transparente para esta etapa):
# - Se faltarem poucos pontos: interpola√ß√£o temporal
# - Caso contr√°rio, fallback para forward fill (ajuste conforme o contexto)
if serie.isna().sum() > 0:
    serie_interp = serie.interpolate(method="time")
    # se ainda houver NaN nas bordas
    serie_interp = serie_interp.ffill().bfill()
else:
    serie_interp = serie.copy()

print("Missing ap√≥s tratamento:", int(serie_interp.isna().sum()))

# Resultado final: s√©rie temporal pronta para an√°lise
ts_usuarios_ativos = serie_interp

print("""
Frequ√™ncia, lacunas temporais e tratamento de missing:
    - Inferir frequ√™ncia (mensal, semanal, di√°ria etc.)
    - Se n√£o conseguir inferir, assumimos mensal (MS) por padr√£o do projeto
    - Regulariza a s√©rie para frequ√™ncia alvo
    - Estrat√©gia de preenchimento (simples e transparente para esta etapa):
        - Se faltarem poucos pontos: interpola√ß√£o temporal
        - Caso contr√°rio, fallback para forward fill (ajuste conforme o contexto)
""")
print("Head (12)")
print(ts_usuarios_ativos.head(12))
````

<div style="background-color:#f3f2ee">
    
<font size="4" color='#FF5733'><strong>üöÄ 5.1 ‚Äî Fun√ß√£o auxiliar para o Teste Dickey-Fuller Aumentado</strong></font>

<font size="3" color='#66666'></font></div>

````python
def teste_adf(serie, nome_serie="S√©rie"):
    """
    Executa o Teste Dickey-Fuller Aumentado (ADF) e exibe os resultados
    de forma organizada.
    """
    resultado = adfuller(serie.dropna(), autolag="AIC")
    
    output = {
        "Estat√≠stica ADF": resultado[0],
        "Valor-p": resultado[1],
        "N¬∫ de Lags Utilizados": resultado[2],
        "N¬∫ de Observa√ß√µes": resultado[3]
    }
    
    print(f"Resultado do Teste ADF ‚Äì {nome_serie}\n")
    for k, v in output.items():
        print(f"{k}: {v:.4f}")
        
    print("\nValores Cr√≠ticos:")
    for nivel, valor in resultado[4].items():
        print(f"  {nivel}: {valor:.4f}")
        
    if resultado[1] <= 0.05:
        print("\nConclus√£o: Rejeita-se H0. A s√©rie √© estacion√°ria.")
    else:
        print("\nConclus√£o: N√£o se rejeita H0. A s√©rie N√ÉO √© estacion√°ria.")
````

````python
teste_adf(ts_usuarios_ativos, nome_serie="Usu√°rios Ativos (S√©rie Original)")
print("""
Teste de estacionariedade na s√©rie original
Avaliar formalmente se a s√©rie original atende ao crit√©rio de estacionariedade.
""")
````

<font size="4" color='#CC403E'><strong>üìå Contextualizando:</strong></font>

<font size="2" color='#66666'>

- Ap√≥s a constru√ß√£o do modelo ARIMA baseline, realiza-se um ajuste fino dos par√¢metros para identificar a combina√ß√£o (p, d, q) que melhor representa a estrutura da s√©rie.
- O crit√©rio utilizado para compara√ß√£o entre os modelos √© o **AIC (Akaike Information Criterion)**, que penaliza modelos excessivamente complexos e favorece aqueles que oferecem melhor equil√≠brio entre ajuste e parcim√¥nia.
- O objetivo desta etapa n√£o √© maximizar desempenho preditivo a qualquer custo, mas sim encontrar um modelo estatisticamente bem especificado e interpret√°vel.

<font size="3" color='#66666'></font></div>

````python
# Como a s√©rie j√° est√° diferenciada, d = 0
p = range(0, 4)
d = [0]
q = range(0, 4)

grid_parametros = list(itertools.product(p, d, q))

print(f"Total de combina√ß√µes testadas: {len(grid_parametros)}")
grid_parametros

print("""
Defini√ß√£o do grid de par√¢metros ARIMA
Limitar p e q entre 0 e 3 √© uma boa pr√°tica para s√©ries curtas/m√©dias e evita modelos inst√°veis.
""")
````

````python
resultados = []

for order in grid_parametros:
    try:
        modelo = ARIMA(train, order=order)
        resultado = modelo.fit()
        
        resultados.append({
            "p": order[0],
            "d": order[1],
            "q": order[2],
            "AIC": resultado.aic,
            "BIC": resultado.bic
        })
    except:
        continue

df_resultados = pd.DataFrame(resultados).sort_values("AIC").reset_index(drop=True)

print('Head (10)')
print(df_resultados.head(10))
print("""
Grid search ARIMA com crit√©rio AIC
Avaliar sistematicamente diferentes especifica√ß√µes ARIMA e ranque√°-las pelo AIC.
""")
````

````python
print("Head (5)")
print(df_resultados.head(5))
print("""
Top modelos ARIMA por AIC
- Leitura esperada
    - Menor AIC ‚Üí melhor equil√≠brio ajuste √ó complexidade
    - Modelos muito pr√≥ximos em AIC devem ser avaliados tamb√©m pelo diagn√≥stico dos res√≠duos
""")
````

<div style="background-color:#f3f2ee">
    
<font size="5" color='#CC403E'><strong>‚öôÔ∏è 9. SARIMA ‚Äì Incorpora√ß√£o de Sazonalidade</strong></font>

<font size="2" color='#66666'></font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#CC403E'><strong>üìå Contextualizando:</strong></font>

<font size="2" color='#66666'>

- As an√°lises explorat√≥rias e a decomposi√ß√£o da s√©rie indicaram a presen√ßa de **padr√µes sazonais**, associados √† periodicidade dos dados.
- Para capturar explicitamente esses padr√µes, utiliza-se o modelo **SARIMA** (Seasonal ARIMA), que estende o ARIMA tradicional ao incluir componentes autorregressivos, de diferencia√ß√£o e de m√©dia m√≥vel sazonais.
- O objetivo desta etapa √© avaliar se a incorpora√ß√£o da sazonalidade resulta em um modelo estatisticamente mais adequado e com melhor desempenho preditivo quando comparado ao ARIMA n√£o sazonal.

---

</font></div>

<div style="background-color:#f3f2ee">
    
<font size="4" color='#FF5733'><strong>üöÄ 9.1. Defini√ß√£o da sazonalidade</strong></font>

<font size="3" color='#66666'></font></div>

````python
# Para s√©ries mensais, assume-se sazonalidade anual
periodo_sazonal = 12
print(f"Per√≠odo sazonal adotado: {periodo_sazonal}")
print("Defini√ß√£o da estrutura sazonal")
````

<div style="background-color:#f3f2ee">
    
<font size="4" color='#FF5733'><strong>üöÄ 9.2. Defini√ß√£o do grid SARIMA (restrito)</strong></font>

<font size="3" color='#66666'></font></div>

````python
# Par√¢metros n√£o sazonais (p, d, q)
p = range(0, 3)
d = [0]  # s√©rie j√° diferenciada
q = range(0, 3)

# Par√¢metros sazonais (P, D, Q, s)
P = range(0, 2)
D = [0]
Q = range(0, 2)
s = periodo_sazonal

parametros_sarima = list(itertools.product(p, d, q, P, D, Q))

print(f"Total de combina√ß√µes SARIMA testadas: {len(parametros_sarima)}")

print("""
Grid de par√¢metros SARIMA
- Nesta c√©lula foi definido um grid restrito de par√¢metros para o modelo SARIMA, combinando componentes n√£o sazonais (p, d, q) e sazonais (P, D, Q, s).
- A restri√ß√£o do grid √© uma decis√£o deliberada para:
    - Evitar sobreajuste
    - Reduzir o tempo de execu√ß√£o
    - Priorizar modelos estatisticamente interpret√°veis
- Como a s√©rie j√° foi diferenciada anteriormente, os par√¢metros `d` e `D` foram mantidos em zero, concentrando a busca nos componentes autorregressivos e de m√©dia m√≥vel.
""")
````

<div style="background-color:#f3f2ee">
    
<font size="4" color='#FF5733'><strong>üöÄ 9.3. Grid search SARIMA com AIC</strong></font>

<font size="3" color='#66666'></font></div>

````python
resultados_sarima = []

for params in parametros_sarima:
    try:
        order = (params[0], params[1], params[2])
        seasonal_order = (params[3], params[4], params[5], s)
        
        modelo = SARIMAX(
            train,
            order=order,
            seasonal_order=seasonal_order,
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        
        resultado = modelo.fit(disp=False)
        
        resultados_sarima.append({
            "order": order,
            "seasonal_order": seasonal_order,
            "AIC": resultado.aic,
            "BIC": resultado.bic
        })
    except:
        continue

df_sarima = (
    pd.DataFrame(resultados_sarima)
    .sort_values("AIC")
    .reset_index(drop=True)
)

print("Head (5)")
print(df_sarima.head(5))
print("""
Grid search SARIMA com crit√©rio AIC
- Nesta etapa foi executado um grid search sistem√°tico sobre as combina√ß√µes definidas anteriormente, utilizando o "AIC (Akaike Information Criterion)" como m√©trica de compara√ß√£o entre os modelos.
- O AIC permite avaliar o equil√≠brio entre:
    - Qualidade do ajuste do modelo
    - Complexidade da especifica√ß√£o
- Os modelos foram ranqueados do menor para o maior AIC, de forma que as primeiras linhas da tabela representam as especifica√ß√µes estatisticamente mais promissoras.
""")
````

<div style="background-color:#f3f2ee">
    
<font size="4" color='#FF5733'><strong>üöÄ 9.4. Sele√ß√£o e ajuste do melhor SARIMA</strong></font>

<font size="3" color='#66666'></font></div>

````python
best_sarima = df_sarima.loc[0]

print("Melhor configura√ß√£o SARIMA:")
print("Order:", best_sarima["order"])
print("Seasonal Order:", best_sarima["seasonal_order"])

modelo_sarima_otimo = SARIMAX(
    train,
    order=best_sarima["order"],
    seasonal_order=best_sarima["seasonal_order"],
    enforce_stationarity=False,
    enforce_invertibility=False
)

resultado_sarima_otimo = modelo_sarima_otimo.fit(disp=False)

print(resultado_sarima_otimo.summary())
print("""
Ajuste do melhor modelo SARIMA
- Com base no menor valor de AIC, foi selecionada a melhor configura√ß√£o SARIMA.
- O modelo foi ent√£o ajustado utilizando apenas o conjunto de treino.
- O resumo estat√≠stico permite avaliar:
    - Signific√¢ncia dos coeficientes
    - Adequa√ß√£o dos termos sazonais
    - Crit√©rios de informa√ß√£o (AIC, BIC)
- Essas informa√ß√µes s√£o fundamentais para validar se a complexidade adicional introduzida pela sazonalidade √© justificada.
""")
````
